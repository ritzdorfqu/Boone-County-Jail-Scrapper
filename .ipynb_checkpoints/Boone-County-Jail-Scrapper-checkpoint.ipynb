```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(r.text)
```

```python
type(soup)
```

```python
table = soup.find('table', class_= 'collapse data-table shadow responsive')
```

```python
type(table)
```

```python
table
```

```python
table.find_all('th')
```

```python
th_all = table.find('thead').find_all('tr')[1].find_all('th')
```

```python
th_all
```

```python
type(th_all)
```

```python
for th in th_all:
    print(th.text)
```

```python
headers = []
```

```python
for th in th_all:
    header = th.text.strip().replace(' ', '_').lower()
    headers.append(header)
```

```python
headers
```

```python
tr_all = table.find_all('tr')[1:]
```

```python
for tr in tr_all:
    for td in tr.find_all('td'):
        print(td.text.strip())
    print('---------------')
```
